```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt

#Plotly is a powerful and versatile Python library used for 
#creating interactive and visually appealing plots and charts

import plotly.graph_objects as go

import seaborn as sns

import plotly.express as px

%matplotlib inline

# reading files 
df1 = pd.read_csv('Unemployment in India.csv')

df2 = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')

df1.head()

df2.head()

df1.info()

df2.info()

df1.shape

df2.shape

df1.describe()

df2.describe()

df1.isnull().sum()

df2.isnull().sum()

# dropping all null values
df1 = df1.dropna()

df1.isnull().sum()

df1.duplicated().sum()

df2.duplicated().sum()

# renaming the columns of the dataset for better understanding
df1.columns = ['State', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate', 'Area']

df2.columns = ['State', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate', 'Region', 'Longitude', 'Latitude']

df1.head()

df2.head()

# State with the highest level of unemployment
df1['State'].value_counts().idxmax()

df1['State'].value_counts()

# State with lowest level of unemployment
df1['State'].value_counts().idxmin()

df2['State'].value_counts().idxmax()

df2['State'].value_counts().idxmin()

import datetime as dt

import calendar

df1['Date'] = pd.to_datetime(df1['Date'], dayfirst = True)
df1['month_int'] = df1['Date'].dt.month
df1['month'] = df1['month_int'].apply(lambda x: calendar.month_abbr[x])

df2['Date'] = pd.to_datetime(df2['Date'], dayfirst = True)
df2['month_int'] = df2['Date'].dt.month
df2['month'] = df2['month_int'].apply(lambda x: calendar.month_abbr[x])

df1['month'].value_counts().idxmax()

df2['month'].value_counts().idxmax()

df1['month'].value_counts().idxmin()

df2['month'].value_counts().idxmin()

df1.head()

df2.head()

df1.drop(columns = ['Frequency', 'month_int'])

df2.drop(columns = ['Frequency', 'month_int'])

# finding top 10 states with highest level of unemployment
df_1 = df1[['State', 'Estimated Unemployment Rate']].groupby('State').sum().sort_values('Estimated Unemployment Rate', ascending = False)
```

```python
df_2 = df2[['State', 'Estimated Unemployment Rate']].groupby('State').sum().sort_values('Estimated Unemployment Rate', ascending = False)
```

```python
# Visualization
fig = plt.figure()
ax0 = fig.add_subplot(1, 2, 1)
df_1[: 10].plot(kind = 'bar', color='orange', figsize=(30, 5), ax = ax0)
ax0.set_title('Top 10 States with highest unemployment')
ax0.set_xlabel('States')
ax0.set_ylabel('Number of people unemployment %')
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(1, 2, 1)
df_2[: 10].plot(kind='bar', color='blue', figsize=(30, 5), ax=ax0)
ax0.set_title('Top 10 States with highest unemployment')
ax0.set_xlabel('States')
ax0.set_ylabel('Number of people unemployment %')
```
